# EMSC2010-W9-L1-NB2

#### The logistic population model:

$x_{\mathrm{next}} = a x_{\mathrm{initial}}(1-x_{\mathrm{initial}})$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size
* $x_{\mathrm{next}}$ is next year's population size
* $a$ is the ability to reproduce.



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
### Code the logistic population model

a = 3.5  # Reproduction ability
x_initial = 0.1 # Initial population size between 0 and 1

num_generations = 10 # Number of generations to simulate
population_history = [x_initial]

print(f"Initial population: {x_initial:.4f}")

for i in range(num_generations):
    x_next = a * x_initial * (1 - x_initial)
    population_history.append(x_next)
    print(f"Generation {i+1}: {x_next:.4f}")
    x_initial = x_next
```

#### The logistic population model with harvesting:

$x_{\mathrm{next}} = a x_{\mathrm{initial}}(1-x_{\mathrm{initial}}) - h$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size.
* $x_{\mathrm{next}}$ is next year's population size.
* $a$ is the ability to reproduce.
* $h$ is the size of the harvest between 0 and 1.


```python
### Code the logistic population model with harvesting

a = 3.5  # Reproduction ability
x_initial = 0.1 # Initial population size between 0 and 1
h = 0.1

num_generations = 10 # Number of generations to simulate
population_history = [x_initial]

print(f"Initial population: {x_initial:.4f}")

for i in range(num_generations):
    x_next = a * x_initial * (1 - x_initial)-h
    population_history.append(x_next)
    print(f"Generation {i+1}: {x_next:.4f}")
    x_initial = x_next
```

#### The logistic population model with randomness:

$x_{\mathrm{next}} = (a+r) x_{\mathrm{initial}}(1-x_{\mathrm{initial}})$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size.
* $x_{\mathrm{next}}$ is next year's population size.
* $a$ is the ability to reproduce.
* $r$ is a normally distributed random number that changes each year.

```python
### Code the logistic population model with randomness
```

```python
a = 3.5  # Reproduction ability
x_initial = 0.1 # Initial population size between 0 and 1
r_std_dev = 0.1 # Standard deviation for random variation 'r'

num_generations = 10 # Number of generations to simulate
population_history_random = [x_initial]

print(f"Initial population: {x_initial:.4f}")

for i in range(num_generations):
    r = np.random.normal(0, r_std_dev) # Random number for this generation
    x_next = (a + r) * x_initial * (1 - x_initial)
    # Ensure x_next stays within [0, 1] range
    x_next = max(0, min(1, x_next))
    population_history_random.append(x_next)
    print(f"Generation {i+1}: a={a:.2f}, r={r:.4f}, x_next={x_next:.4f}")
    x_initial = x_next
```

#### The logistic map:

$x_{\mathrm{next}} = a x_{\mathrm{initial}}(1-x_{\mathrm{initial}})$

where:
* $x$ is the population size between 0 and 1.
* $x_{\mathrm{initial}}$ is the current population size
* $x_{\mathrm{next}}$ is next year's population size
* $a$ is the ability to reproduce.

Follow these steps:

1. Choose a random value of $a$ from 0 to 4.

2. Choose a random starting population x_{\mathrm{initial}} from 0 to 1

3. Run the model for 2,000 generations and record the $x$ value of the last generation.

4. Repeat steps 1-3 a large number of times (thousands)

5. Make a scatter plot of $a$ (x-axis) versus final $x$ (y-axis).



```python

# Parameters for the simulation
num_iterations = 10000  # Number of times to repeat steps 1-3
num_generations_per_iteration = 2000 # Number of generations for each simulation

# Lists to store results for plotting
a_values = []
final_x_values = []

for _ in range(num_iterations):
    # Step 1: Choose a random value of 'a' from 0 to 4
    a = np.random.uniform(0, 4)

    # Step 2: Choose a random starting population x_initial from 0 to 1
    x_initial = np.random.uniform(0, 1)

    # Step 3: Run the model for 2,000 generations and record the x value of the last generation
    x = x_initial
    for _ in range(num_generations_per_iteration):
        x = a * x * (1 - x)

    # Store the results
    a_values.append(a)
    final_x_values.append(x)

# Step 5: Make a scatter plot of a (x-axis) versus final x (y-axis)
plt.figure(figsize=(10, 6))
plt.scatter(a_values, final_x_values, s=0.1, alpha=0.5)
plt.title('Logistic Map Bifurcation Diagram')
plt.xlabel('Parameter a')
plt.ylabel('Final x value (after 2000 generations)')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()
```
